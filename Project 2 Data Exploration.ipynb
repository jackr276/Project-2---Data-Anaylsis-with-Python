Remove the rows that are mostly NA's in both data sets(these are the last few rows)

```python
entry_survey.dropna(subset=["coursename", "degree"], inplace = True)
literacy_survey.dropna(subset=["coursename", "degree"], inplace = True)
entry_survey
```

for each dataframe, replace values for race that contain a ";" with multiracial

```python
entry_survey["race"] = np.where(entry_survey["race"].str.contains(";"), "Multi-Racial", entry_survey["race"])
literacy_survey["race"] = np.where(literacy_survey["race"].str.contains(";"), "Multi-Racial", literacy_survey["race"])
```
